**Plot some histograms**

Write a function ``assignment_03`` creates a normally distributed variable with mean 5 and standard deviation 2. Plot the histogram.


```python
def assignment_03():
    # INSERT CODE HERE
    # parameters: mean (loc) = 5, std deviation (scale) = 2, number of samples = 1000
    data = np.random.normal(5, 2, 10000)
    
    # 'bins' determines how many bars the histogram has. 
    plt.hist(data, bins=10)
    
    plt.show()


assignment_03()
```

## Assignment 04

**Plot some 2D Kernel Density Estimates**

Write a function ``assignment_04`` creates a 2D normally distributed variable with mean [0,0] and covariance matrix [[5, 2], [2, 2]]. Plot the joint kernel density estimate.


```python
def assignment_04():
    # INSERT CODE HERE
    
    np.random.seed(42)
    #the parameters
    mean = [0, 0]
    cov = [[5, 2], [2, 2]]
    
    # Generate 2D normally distributed data
    data = np.random.multivariate_normal(mean, cov, 1000)
    
    # Separate the data into x and y columns for the plot
    x = data[:, 0]
    y = data[:, 1]
    g = sns.jointplot(x=x, y=y, kind="kde")
    
    g.set_axis_labels('x', 'y')
    plt.show()


assignment_04()
```
